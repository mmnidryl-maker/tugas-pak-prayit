4. fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)
def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")
print("total:", jumlahkan(1, 2, 3, 4, 5))
biodata(nama="budi", umur=20, prodi="TRK")
```

```python
def penjumlahan(*args):
  return sum(args)
print("total: ", penjumlahan(1, 2, 3, 4, 5))
```

7. fungsi decorator

```python
def b(c):
    def a():
        print("sebelum fungsi dijalankan...")
        c()
        print("setelah dungsi dijalankan...")
    return a

@b
def sapa():
    print("halo, mahasiswa polines!")

sapa()
```

```python
def b(c):
  def a():
    print("ku kalikan dulu bos")
    c()
    print("udah kukalikan ya bos")
  return a

@b
def kali():
  x = 10*9
  print("hasil kali = ", x)

kali()
```

8. menghitung rata-rata nilai mahasiswa

```python
def hitung_rata(data):
  return sum(data) / len(data)

nilai = (80, 90, 85)
rata = hitung_rata(nilai)

print("rata-rata nilai:", rata)
```

4 fungsi dengan *args

```python
def jumlahkan_semua(*args):
    return sum(args)

angka_list = []
while True:
    user_input = input("masukan angka (ketik selesai untuk menjumlahkan): ")
    if user_input.lower() == "selesai":
        break
    try:
        angka = int(user_input)
        angka_list.append(angka)
    except ValueError:
        print("Input tidak valid. Harap masukkan angka atau 'selesai'.")

print("total:", jumlahkan_semua(*angka_list))
```

11 studi kasus nyata program nilai

```python
nilai_1 = (80, 90, 85)
rata1 = sum(nilai_1) / len(nilai_1)
nilai_2 = (75, 85, 90)
rata2 = sum(nilai_2) / len(nilai_2)

nilai_akhir = lambda rata1, rata2: rata2 * 0.4 + rata1 * 0.6
print("nilai akhir:", nilai_akhir(rata1, rata2))

print("rata-rata nilai:", rata)
if rata >= 80:
  print("lulus")
else:
  print("tidak lulus")
```
