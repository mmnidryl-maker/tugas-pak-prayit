 Modul buatan sendiri

```python
%%writefile konversi.py
def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
import konversi

suhu_c = 30
hasil_f = konversi.celcius_ke_fahrenheit(suhu_c)
print(f"{suhu_c} Celcius = {hasil_f} Fahrenheit")

suhu_f = 100
hasil_c = konversi.fahrenheit_ke_celcius(suhu_f)
print(f"{suhu_f} Fahrenheit = {hasil_c:.2f} Celcius")
```

Studi Kasus Mini

```python
%%writefile laporan.py

def tampilkan_hasil(benar, jumlah_soal):
    nilai = (benar / jumlah_soal) * 100

    print("\n--- HASIL KUIS ---")
    print(f"Skor Anda: {nilai:.2f}%")
```

```python
import random
import laporan

jumlah_soal = 3
skor = 0

for i in range(jumlah_soal):
    angka1 = random.randint(1, 10)
    angka2 = random.randint(1, 10)
    soal = f"{angka1} + {angka2} = "
    jawaban = int(input(soal))

    if jawaban == angka1 + angka2:
        print("Jawaban Anda Benar!")
        skor += 1

    else:
      print("Jawaban Anda Salah!")

laporan.tampilkan_hasil(skor, jumlah_soal)
```
